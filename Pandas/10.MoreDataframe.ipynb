## astype() method

```python
cast = pd.DataFrame(
    {
        'name': ['Afghanistan', 'Albania', 'Algeria', 'Andorra', 'Angola'],
        'area': [652230, 28748, 2381741, 468, 1246700],
        'population': ['2550010.0 million', '283174.1 million', '3710000.0 million', '7811.5 million', '2060929.4 million'],
        'gdp': ['8.9' , '7.4', '1.3', '2.5', '7.5']
    }
)
```

```python
cast
```

```python
cast.info()
```

```python
cast = cast.astype({'area': int,
                    'gdp': float})
```

```python
cast
```

```python
cast.info()
```

```python
cast.name.astype(float)
```

Does not work with actual strings or mixed numeric strings

```python
data = pd.DataFrame(data)
data
```
